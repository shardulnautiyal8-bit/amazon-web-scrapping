```python
response=requests.get(url,headers=header)
```

```python
if response.status_code==200:
    html_content=response.text
else:
    print("error")
```

```python
soup=bs(html_content,'html.parser')
```

```python
product_name = soup.find("span" , id = "productTitle").text.strip()
```

```python
print(product_name)
```

```python
product_price = soup.find("span" , class_ = "a-price-whole").text.strip()
```

```python
print(product_price)
```

```python
about_this_item = soup.find("ul" , class_ = "a-unordered-list a-vertical a-spacing-mini").text.strip()
```

```python
print(about_this_item)
```

```python
customer_say=soup.find("span",class_= "aui-primitive __SAR2l0zNyyuZ").text.strip()
```

```python
print(customer_say)
```

```python
final_review=soup.find("span",class_= "a-size-medium a-color-base").text.strip()
```

```python
print(final_review)
```

```python
with open("output.csv", mode = 'w', newline = "", encoding = "utf-8") as file :
    writer = csv.writer (file)
    writer. writerow(["product_name", "product_price", 'about_this_item', "customer_say", "final_review" ])
    writer. writerow([product_name,product_price, about_this_item, customer_say, final_review])

print('data saved')
```
